```python
import operator
import mimetypes, urllib2
# get top full URLS

sorted_image_urls = sorted(full_urls_count.items(), key=operator.itemgetter(0), reverse=True)
# print(sorted_image_urls)


# Find it image is in URL
# https://stackoverflow.com/questions/10543940/check-if-a-url-to-an-image-is-up-and-exists-in-python


def is_url_image(url):    
    mimetype,encoding = mimetypes.guess_type(url)
    return (mimetype and mimetype.startswith('image'))

def check_url(url):
    """Returns True if the url returns a response code between 200-300,
       otherwise return False.
    """
    try:
        headers={
            "Range": "bytes=0-10",
            "User-Agent": "MyTestAgent",
            "Accept":"*/*"
        }

        req = urllib2.Request(url, headers=headers)
        response = urllib2.urlopen(req)
        return response.code in range(200, 209)
    except Exception, ex:
        return False

def is_image_and_ready(url):
    return is_url_image(url) and check_url(url)

def download_images(urls, save_path):
    for url in urls:
        file_name, mime = urllib.urlretrieve(url)
        name = url.split('/')[-1]
#         print url
        shutil.copy(file_name, save_path + name)

        
for url, count in sorted_image_urls:
    if( is_image_and_ready(url)):
        print(url)
        download_images([url], '/media/sumeet/hd2/sumeetstuff/share/black_panther/images/')
        
```
